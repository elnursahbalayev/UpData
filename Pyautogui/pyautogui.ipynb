```python
# Spiral using Puautogui
time.sleep(3)

distance = 300
while distance > 0:
    pa.dragRel(distance, 0, button='left')
    pa.dragRel(0, distance, button='left')
    distance -= 20
    pa.dragRel(-distance, 0, button='left')
    pa.dragRel(0, -distance, button='left')
    distance -= 20
```
