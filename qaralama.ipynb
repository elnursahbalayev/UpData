```python
import random
import seaborn as sns
import numpy as np
```

```python
uniform_distribution = np.random.uniform(1,10,10000)

sns.histplot(uniform_distribution)
```

```python
uniform_distribution = np.random.uniform(1,10,1000000)

sns.histplot(uniform_distribution)
```

```python
averages = []
```

```python
for i in range(1000):
    mean_value = np.mean(random.choices(uniform_distribution,k=100))
    averages.append(mean_value)
```

```python
sns.histplot(averages)
```
