## Asymmetry - skewness

Skewness adından da göründüyü kimi həm də asymmetry adlanır. Bu isə deməkdir ki simmetriyanın olmadığı distribution-ları göstərir. Skewness simmetrik bell curve-dən vəya normal distribution-dan kənara çıxmanı göstərir.

### What is bell curve / normal distribution?

bell curve.svg

skewness.png

```python
# kitabxanaların daxil olunması
import numpy as np
import random

# 1 və 100 arasında 100000 (yüz min) random ədəd götürmək
random_ededler = np.random.normal(500, 500,100000)
random_ededler
```

```python
# neçə random ədədimiz olduğuna baxmaq
len(random_ededler)
```

```python
import seaborn as sns

sns.distplot(random_ededler)
```

```python
np.mean(random_ededler)
```

```python
np.median(random_ededler)
```

## Box plots, quartiles and percentiles, interquartile range

```python
import seaborn as sns

df_iris = sns.load_dataset('iris')

df_iris.head(10)
```

```python
sns.boxplot(x=df_iris['sepal_length'])
```

```python
import plotly.express as px
df = px.data.tips()
fig = px.box(df, y='total_bill')
fig.show()
```

```python
fig = px.box(df, x='time', y='total_bill')
fig.show()
```
