```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import time
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.action_chains import ActionChains

import pandas as pd
```

```python
PATH = "C:\Program Files (x86)\chromedriver.exe"
driver = webdriver.Chrome(PATH)
driver.maximize_window()
```

```python
df = pd.read_csv('Car_links_turboaz.csv')
df
```

```python
elan_nomresi_list,seher, marka, model, buraxilis_ili,ban_novu, reng, muherrik, muherrik_gucu, yanacaq_novu, yurus, suretler_qutusu, oturucu, yeni, qiymet, yerlerin_sayi, sahibler, veziyyeti = [],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[]


for i in range(len(df['Links'])//10*6,len(df['Links'])):
    try:
        driver.get(df['Links'][i])
        elan_nomresi = driver.find_element_by_class_name('product-statistics')
        elan_nomresi = int(elan_nomresi.text.split()[-1])
        elan_nomresi_list.append(elan_nomresi)
        properties = driver.find_elements_by_class_name('product-properties-value')

        seher.append(properties[0].text)
        marka.append(properties[1].text)
        model.append(properties[2].text)
        buraxilis_ili.append(properties[3].text)
        ban_novu.append(properties[4].text)
        reng.append(properties[5].text)
        muherrik.append(properties[6].text)
        muherrik_gucu.append(properties[7].text)
        yanacaq_novu.append(properties[8].text)
        yurus.append(properties[9].text)
        suretler_qutusu.append(properties[10].text)
        oturucu.append(properties[11].text)
        yeni.append(properties[12].text)
        qiymet.append(properties[13].text)
        # yerlerin_sayi.append(properties[14].text)
        # sahibler.append(properties[15].text)
        # veziyyeti.append(properties[16].text)
    except:
        continue
```
