```python
x1 = [1,3, 5, 7]
y1 = [2,4, 6, 8]

x2 = [7,7,7,7]
y2 = [17, 18, 29, 40]

plt.figure(figsize=(10,5))

plt.bar(x1, y1, label='First')
plt.bar(x2, y2, label = 'Second')

plt.xlabel('X')
plt.ylabel('Y')

plt.title('X vs Y')

plt.legend()
```

```python
x1 = [1,3, 5, 7]
y1 = [2,4, 6, 8]

x2 = [7,7,7,7]
y2 = [17, 18, 29, 40]

plt.figure(figsize=(10,5))

plt.barh(x1, y1, label='First')
plt.barh(x2, y2, label = 'Second')

plt.xlabel('X')
plt.ylabel('Y')

plt.title('X vs Y')

plt.legend()
```

```python
plt.figure(figsize=(8,8))

x = np.arange(1,10)
y = np.random.randint(100,1001,9)

plt.bar(x,y)
plt.title('X vs Y')

for i in range(len(x)):
    plt.text(x[i], y[i] + 10, s=f'{y[i]}', ha='center')


plt.xlabel('Employees')
plt.ylabel('Salary')
```

```python
x1 = ['Sachin', 'Savin', 'Saurav', 'Rahul']
y1 = [17,18, 459, 20]
y2 = [21, 23, 25, 28]

plt.figure(figsize=(7,5))

plt.bar(x1, y1, label='y1', width=0.5)
plt.bar(x1, y2, label='y2', width=0.5)

plt.legend()
```

## Line graph

```python
x = np.linspace(0,10,1000)
y = np.sin(x)

plt.plot(x,y)

plt.show()
```

```python
x = np.linspace(0,10,100)
y = np.sin(x)
y2 = np.cos(x)

plt.plot(x,y, 'r--o')
plt.plot(x,y2,'g-*')

plt.tight_layout()
plt.show()
```

```python
x = np.linspace(0,10,100)
y = np.sin(x)
y2 = np.cos(x)

plt.plot(x,y,  linestyle='--')
plt.plot(x,y2,'g-*', linewidth=2)

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(15,5))

x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.subplot(1,2,1)
plt.plot(x,y1)
plt.subplot(1,2,2)
plt.plot(x,y2)
```

```python
x = np.arange(1,10)

y = np.array([x[:i+1].sum() for i in range(len(x))])
```

```python
y
```

```python
fig, ax = plt.subplots(2,2, figsize=(12,12))

ax[0,0].plot(x,y)
ax[0,1].bar(x,y)
ax[1,0].plot(x,y**2)
ax[1,1].bar(x, y**2)
```

## Scatter plot

```python
x = np.random.randint(0,100, 100)
y = np.random.randint(0,1000, 100)


fig, ax = plt.subplots(1,2, figsize=(10,5))

ax[0].plot(x, y,color='b')
ax[1].scatter(x,y, color='r')

fig.savefig('sekil 1.png')
```

```python
plt.scatter(x,y,s=50)
plt.tight_layout()
plt.xlabel('xtreme positions', fontsize=15, fontweight='bold')
```

## Pie chart/

```python
area = [48, 32, 20, 15]
labels = ['a', 'c', 'd' ,'e']

plt.figure()
plt.pie(area ,labels=labels, startangle=45, shadow=True, autopct='%.0f%%')

plt.show()
```

## Histogram

```python
x = np.random.normal(0,10,10000)

plt.hist(x, bins=40)
plt.show()
```

```python
x = np.random.uniform(0,10,10000)

plt.hist(x, bins=40)
plt.show()
```

```python
averages = []
for i in range(1000):
    averages.append((np.random.choice(x, size=50)).mean())
    
averages = np.array(averages)

plt.hist(averages)
```
