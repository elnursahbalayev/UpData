# 5 Bar Graph  <a class="anchor" id="5"></a>

[  ](#0.1)

Bar graph əsasən iki kateqoriya və ya qruplar arasında müqayisə etmək üçün istifadə olunur. Məsələn, işçilərin və onların maaşlarının müqayisəsini göstərmək istəyirsiniz.

```python
## Basic Plot
plt.figure(figsize=(10,10))
id1 = np.arange(1,10)
salary = np.arange(20,110,10)
plt.bar(id1,salary)
plt.xlabel('Employee ID')
plt.ylabel('Salary')
plt.show()
```

**Biz aşağıdakı funksiyalarla qrafiklərinizi stilələyə bilərik**
1. **plt.title( )** qrafiklərə başlıq təyin etmək üçün.
2. **plt.xlabel( )** x-oxunu adlandırmaq üçün.
3. **plt.ylabel( )** y-oxunu adlandırmaq üçün.
4. **color** = seçimini plt.bar( ) barların rəngini təyin etmək üçün.

```python
id1 = np.arange(1,10)
salary = np.arange(20,110,10)
plt.figure(figsize=(8,5)) # fiqurun ölçüsü 
plt.bar(id1,salary,color = '#FFA726')
plt.xlabel(r'$Employee $ $ ID$')
plt.ylabel(r'$Salary$')
plt.show()
```

```python
#bir neçə seti bir arada göstərmək
x1= [1,3,5,7]
x2=[2,4,6,8]
y1 = [7,7,7,7]
y2= [17,18,29,40]
plt.figure(figsize=(8,6))
plt.bar(x1,y1,label = "First",color = '#42B300') # birinci set
plt.bar(x2,y2,label = "Second",color = '#94E413') # ikinci set
plt.xlabel('$X$')
plt.ylabel('$Y$')
plt.title ('$Bar $ $ Chart$')
plt.legend()
plt.show()
```

## Stacked Vertical Bar  <a class="anchor" id="5.7"></a>

[  ](#0.1)

```python
x1= ['Sachin','Saurav','Rahul','Robin']
y1= [17,18,29,40]
y2 = [20,21,22,23]
plt.figure(figsize=(5,7))
plt.bar(x1,y1,label = "y1",width = 0.5)
plt.bar(x1,y2,label = "y2",width = 0.5 ,bottom = y1 )
plt.xlabel('$X$')
plt.ylabel('$Y$')
plt.title ('$Bar $ $ Chart$')
plt.legend()
plt.show()
```

```python

plt.figure(figsize=(7,9))
# bar-ın eni
barWidth = 0.25
# bar-ın hündürlüyü
y1= np.array([17,18,29,40])
y2 =np.array([20,21,22,23])
y3 =np.array([5,9,11,12])
# pozisiyalar
pos1 = np.arange(len(y1))
pos2 = [x + barWidth for x in pos1]
pos3 = [x + barWidth for x in pos2]
# plot-u göstərmək
plt.bar(pos1, y1, color='#FBC02D', width=barWidth, label='Open')
plt.bar(pos2, y2, color='#F57F17', width=barWidth, label='Closed')
plt.bar(pos3, y3, color='#E65100', width=barWidth, label='Cancelled')

# xtick-lərin əlavə edilməsi
plt.xlabel('Assignee', fontweight='bold')
plt.ylabel('Number of Tickets', fontweight='bold')
plt.xticks([i + barWidth for i in range(len(y1))], ['Asif', 'Basit', 'Ravi', 'Minil'])
for i in range(len(pos1)):
    plt.text(pos1[i], y1[i]+0.5, s=f'{y1[i]}', ha='center')
for i in range(len(pos2)):
    plt.text(pos2[i], y2[i]+0.5, s=f'{y2[i]}', ha='center')
for i in range(len(pos3)):
    plt.text(pos3[i], y3[i]+0.5, s=f'{y3[i]}', ha='center')
plt.title ('$Grouped $ $ Bar $ $ Chart$')

# legend-ə əlavə edilməsi
plt.legend()
# göstərmək
plt.show()
```

## 6.2 Bir neçə xəttinn əlavə edilməsi


```python
plt.figure(figsize=(10,5))
x = np.array([1,2,3,4,5,6])
y1 = np.power(x,2)
y2 = np.power(x,3)
plt.plot(x,y1, "b-" , label = '$y1 = x^2$') # Setting up legends
plt.plot(x,y2, "r-" ,label ='$y2 = x^3$') # Setting up legends
plt.xlabel("X - Axis")
plt.ylabel("Y - Axis")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
#bir neçə setin göstərilməsi
x = np.linspace(0, 10, 2000)
plt.figure(figsize=(10,6))
plt.plot(x,np.sin(x) , label = '$Sin(X)$')
plt.plot(x,np.cos(x) , label = '$cos(X)$')
plt.xlabel(r'$X$' , fontsize = 18)
plt.ylabel(r'$Y$' , fontsize = 18)
plt.title("$Sin(x) $ $ & $ $ Cos(x)$" ,fontsize = 14)
plt.legend(loc = 'upper right') # legend pozisiya
plt.show()
```

```python
# # Display multiple plots in one figure using subplots()
x = np.arange(-50,50)
y1 = np.power(x,2)
y2 = np.power(x,3)
y3 = np.sin(x)
y4 = np.cos(x)
y5 = np.tan(x)
y6 = np.tanh(x)
y7 = np.sinh(x)
y8 = np.cosh(x)
y9 = np.exp(x)
fig1 , ax1 = plt.subplots(nrows=3,ncols=3 , figsize = (20,20)) # Create a figure and subplo
ax1[0,0].plot(x,y1,"tab:blue") # set the color of the line chart
ax1[0,0].set_title("Square Function") # setting title of subplot
ax1[0,0].set_xlabel(r'$X$' , fontsize = 18) #Set the label for the x-axis
ax1[0,0].set_ylabel(r'$Y$' , fontsize = 18) #Set the label for the y-axis
ax1[0,1].plot(x,y2,"tab:orange")
ax1[0,1].set_title("Cubic Function")
ax1[0,1].set_xlabel(r'$X$' , fontsize = 18)
ax1[0,1].set_ylabel(r'$Y$' , fontsize = 18)
ax1[0,2].plot(x,y3,"tab:green")
ax1[0,2].set_title("Sine Function")
ax1[0,2].set_xlabel(r'$X$' , fontsize = 18)
ax1[0,2].set_ylabel(r'$Y$' , fontsize = 18)
ax1[1,0].plot(x,y4,"b-")
ax1[1,0].set_title("Cosine Function")
ax1[1,0].set_xlabel(r'$X$' , fontsize = 18)
ax1[1,0].set_ylabel(r'$Y$' , fontsize = 18)
ax1[1,1].plot(x,y5,"r-")
ax1[1,1].set_title("Tangent Function")
ax1[1,1].set_xlabel(r'$X$' , fontsize = 18)
ax1[1,1].set_ylabel(r'$Y$' , fontsize = 18)
ax1[1,2].plot(x,y6,"g-")
ax1[1,2].set_title("Hyperbolic Tangent")
ax1[1,2].set_xlabel(r'$X$' , fontsize = 18)
ax1[1,2].set_ylabel(r'$Y$' , fontsize = 18)

ax1[2,0].plot(x,y7,"m-")
ax1[2,0].set_title("Hyperbolic Sine")
ax1[2,0].set_xlabel(r'$X$' , fontsize = 18)
ax1[2,0].set_ylabel(r'$Y$' , fontsize = 18)
ax1[2,1].plot(x,y8,"y-")
ax1[2,1].set_title("Hyperbolic Cosine")
ax1[2,1].set_xlabel(r'$X$' , fontsize = 18)
ax1[2,1].set_ylabel(r'$Y$' , fontsize = 18)
ax1[2,2].plot(x,y9,"k-")

ax1[2,2].set_title("Exponential Function")
ax1[2,2].set_xlabel(r'$X$' , fontsize = 18)
ax1[2,2].set_ylabel(r'$Y$' , fontsize = 18)
plt.show()
```

## 8 Pie Chart  <a class="anchor" id="8"></a>

[  ](#0.1)


Pie chart dairəvi grafikdir, məlumatı kateqoriyaların hər birinin nisbətini göstərmək üçün dilimlərə bölmək üçün istifadə olunur. Əgər siz faizləri göstərmirsinizsə, hər biri 100% ə ehtiva etməlidir.

```python
plt.figure(figsize=(9,9))
area = [48 , 30 , 20 , 15]
labels = ['Low' , 'Medium' , 'High' , 'Critical']
plt.pie (area , labels= labels)
plt.show()
```

```python
#rənglərə görə bölmək
plt.figure(figsize=(9,9))
area = [48 , 30 , 20 , 15]
labels = ['Low' , 'Medium' , 'High' , 'Critical']
colors = ['#8BC34A','#D4E157','#FFB300','#FF7043']
plt.pie (area , labels= labels , colors= colors)
plt.show()
```

## Faiz və qiymətin göstərilməsi   <a class="anchor" id="8.1"></a>

[  ](#0.1)

```python
# Display percentage in Pie Chart using autopct='%1.1f%%'
plt.figure(figsize=(8,8))
area = [48 , 30 , 20 , 15]
labels = ['Low' , 'Medium' , 'High' , 'Critical']
colors = ['#7CB342','#C0CA33','#FFB300','#F57C00']
plt.pie (area , labels= labels , colors= colors , startangle=45 , shadow='true', autopct='%.1f%%')
plt.show()
```

```python
plt.figure(figsize=(8,8))
area = [48 , 30 , 20 , 15]
total = np.sum(area)
labels = ['Low' , 'Medium' , 'High' , 'Critical']
def val_per(x):
    return f'{x/total*100:.2f}%\n{x:.0f}'
colors = ['#7CB342','#C0CA33','#FFB300','#F57C00']
plt.pie (area , labels= labels , colors= colors , startangle=45 , shadow='true', autopct=val_per)
plt.show()
```

##  3 D Plotlar

3D plotlar əsasən 3-cü müstəvidə vizuallaşdırma zamanı istifadə olunur

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = Axes3D(fig)
X = np.arange(-4, 4, 0.25)
Y = np.arange(-4, 4, 0.25)
X, Y = np.meshgrid(X, Y)
Z = X**2 + Y**2
ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap='hot')
```
